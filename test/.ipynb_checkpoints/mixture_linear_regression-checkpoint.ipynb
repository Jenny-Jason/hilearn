# Study of Mixture linear regression

This notebook shows an implementation of mixture of linear regression, and some test study.

## Linear regression: Gaussian noise and least squares

In the linear regression model, we assume that the dependent variable $y$ is a linear product of independent variabl $\textbf{x}$ and the weight $\textbf{w}$ with additive Gaussian noise, so that,

$y=\textbf{w}^\top \cdot \textbf{x} + \epsilon$;         $\epsilon \sim \mathcal{N}(0, \sigma)$

The probability or the likelihood of this model on a set of observation {$\textbf{X}, Y$} is follow,

$p(Y|\textbf{X}, \textbf{w}, \sigma) = \prod_{n=1}^{N} \mathcal{N}(y_n|\textbf{w}^\top \cdot \textbf{x}_n, \sigma)$

Then the log likelihood:

$\begin{split}
\log p(Y|\textbf{X}, \textbf{w}, \sigma) & = \sum_{n=1}^{N} \log \mathcal{N}(y_n|\textbf{x}_n \cdot \textbf{w}, \sigma) \\
& = -N\log \sigma -\frac N 2 \log 2\pi - \frac 1 {2\sigma^2} \sum_{n=1}^{N} (y_n - \textbf{w}^\top \cdot \textbf{x}_n)^2
\end{split}$

Given the variance of Gaussian noise, maximizing the likelihood is the same as minizing the squares errors.

$\triangledown \log p(Y|\textbf{X}, \textbf{w}, \sigma) = \frac 1 {\sigma^2} \sum_{n=1}^{N} \lbrace (y_n - \textbf{w}^\top \cdot \textbf{x}_n) \cdot \textbf{x}_n^\top \rbrace = 0
\Rightarrow \sum_{n=1}^{N} y_n \textbf{x}_n^\top - \textbf{w}^\top \sum_{n=1}^{N} \textbf{x}_n \textbf{x}_n^\top = 0$

Solving it, we will have,

$\textbf{w}_{ML} =  (\textbf{X}^\top \textbf{X})^{-1} \textbf{X}^\top Y$

Note: in some cases, e.g. lack of data, $\textbf{X}^\top \textbf{X}$ may be not inversible, which means there are multiple perfect solutions.

```python
import numpy as np
    
class LinearRegression:
    def __init__(self):
        """an initialization of the model, which could be extended
        with initial settings."""
        pass
    def fit(self, X, Y):
        """implement the training"""
        # check the shape of the input
        if len(X.shape) == 1:
            X = X.reshape(-1,1)
        if len(Y.shape) == 2 and Y.shape[1] == 1:
            Y = Y.reshape(-1)
        self.X = X
        self.Y = Y
        N = len(Y)
        X1 = np.ones((X.shape[0], X.shape[1]+1))
        X1[:,1:] = X
        coef = np.linalg.inv(np.dot(X1.T, X1)).dot(X1.T).dot(Y)
        #coef = np.linalg.lstsq(X1, Y)[0]
        sigma = np.std(Y - X1.dot(coef))
        loglik = -N*np.log(sigma) - 0.5*N*np.log(2*np.pi) - np.sum((Y - X1.dot(coef))**2) / (2*sigma**2)
        
        self.coef_ = coef[1:]
        self.intercept_ = coef[0]
        self.loglik = loglik
        self.sigma  = sigma
        
    def predict(self, Xtest):
        """implement the prediction"""
        if len(Xtest.shape) == 1:
            Xtest = Xtest.reshape(-1,1)
        RV = Xtest.dot(self.coef_) + self.intercept_
        return RV.reshape(-1)
```

```python
import scipy.stats as st
import pylab as pl
%matplotlib inline

X = np.random.rand(200, 5)
W = np.random.rand(5) * 20
Y = X.dot(W) + np.random.rand(200) * 5

# un-inversible
# X = np.array([[0, 0], [1, 1], [2, 2]])
# Y = np.array([0, 1, 2])

linreg = LinearRegression()
linreg.fit(X,Y)
Y_pre = linreg.predict(X)
print linreg.coef_, linreg.intercept_, linreg.sigma
print W, np.mean(Y-X.dot(W)), np.std(Y-X.dot(W))

fig = pl.figure()
pl.plot(Y, Y_pre, "o", alpha=0.5)
pl.xlabel("observed $y$")
pl.ylabel("predicted $y$")
pl.title("Pearson's $R=%.3f$" %(st.pearsonr(Y, Y_pre)[0]))

# fig.set_size_inches(5,4)
# fig.savefig(data_dir = "test1.pdf", dpi=300, bbox_inches='tight')
pl.show()
```

## Mixture of linear regression

Sometimes the collected data can be explained well just by a single linear regression model. There maybe a mixture of multiple linear regression models. Therefore learning a mixed architecture can be very useful.

Suppose that we have $K$ components in the data, and the prior array for these components is $\pi$, then the posterior of each data point is in the following form,

$p(y_i|\textbf{x}_i, \pi, \theta) = \sum_{k=1}^K \pi_k \mathcal{N}(y_i|\textbf{w}_k^\top \cdot \textbf{x}_i, \sigma_k)$,

where $\sum_{k=1}^K \pi_k = 1$ as prior array, and $\theta=\{\textbf{w}_1, \sigma_1, ..., \textbf{w}_K, \sigma_K \}$ is a set of component sepecific parameters. The log likelihood of full data set is follow,

$p(Y|\textbf{X}, \pi, \theta) = 
\sum_{n=1}^N \log \sum_{k=1}^K \pi_k \mathcal{N}(y_n|\textbf{w}_k^\top \cdot \textbf{x}_n, \sigma_k) = 
\sum_{n=1}^N \sum_{k=1}^K z_{nk} \log \pi_k \mathcal{N}(y_n|\textbf{w}_k^\top \cdot \textbf{x}_n, \sigma_k)$

where, we introduced a set of letant binary variables $\textbf{Z}=\{z_n\}$ as the identities of components, where $z_{nk}\in\{0,1\}$, and $\sum_{k=1}^K z_{nk} = 1$. However, the maximum likelihood cannot be achieved analytically. Therefore, we introduce an EM algorithm to estimate the paramters. 

####Random initialization
Randomly set hard partition of the $N$ points into $K$ components and calculate the parameters

####E-step
Based on the parameters $(\mathbf\pi, \mathbf\theta)$, calculate the expectation of the componenet identities $(I_n=k)$, as follow,

$r_{nk} = \mathbb{E}(I_n=k)= p(I_n=k|y_n,\textbf{x}_n,\pi^{\text{old}},\theta^{\text{old}})=\frac 
{\pi_k \mathcal{N}(y_n|\textbf{w}_k^\top \cdot \textbf{x}_n, \sigma_k)} 
{\sum_{j=1}^K \pi_j \mathcal{N}(y_n|\textbf{w}_j^\top \cdot \textbf{x}_n, \sigma_j)}$

####M-step
Based on the expectation of the identities, the log likelihood will be maximized, as follow

$Q(\pi,\theta|\pi^\text{old}, \theta^\text{old})=\sum_{n=1}^N \sum_{k=1}^K r_{nk} \log \pi_k \mathcal{N}(y_n|\textbf{w}_k^\top \cdot \textbf{x}_n, \sigma_k)$

Having the derivation as zero, we will have

$\frac {\partial Q} {\partial \pi_k} = 0 \Rightarrow \pi_k = \frac 1 N \sum_{n=1}^N r_{nk}$ 
(accounting $\sum_{k=1}^K z_{nk} = 1$)

----

$\frac {\partial Q} {\partial \textbf {w}_k} = \frac 1 {2\sigma_k^2} \sum_{n=1}^{N} \lbrace r_{nk} (y_n - \textbf{w}_k^\top \cdot \textbf{x}_n) \cdot \textbf{x}_n^\top \rbrace = 0 \Rightarrow
\sum_{n=1}^{N} r_{nk} y_n \textbf{x}_n^\top - \textbf{w}^\top \sum_{n=1}^{N} \textbf{x}_n r_{nk} \textbf{x}_n^\top = 0
\Rightarrow$

$\textbf{w}_{k} =  (\textbf{X}^\top R_{k} \textbf{X})^{-1} \textbf{X}^\top R_{k} Y$

----

$\frac {\partial Q} {\partial \sigma_k} = - \frac {\sum_{n=1}^N r_{nk}} {\sigma_k} + \sum_{n=1}^N \lbrace r_{nk} \frac 1 {\sigma_k^3} (y_n - \textbf{w}^\top \textbf{x})^2\rbrace=0 \Rightarrow$

$\sigma_k = \sqrt {\frac 1 {\sum_{n=1}^N r_{nk}} \sum_{n=1}^N \lbrace r_{nk} (y_n - \textbf{w}^\top \textbf{x})^2\rbrace}$


In this way, the full likelihood will increase during iteration, however, there are still some drawback, especially local optimum and overfitting. For the former, we could run multiple chains and choose the best one; for the latler, we could try bayes information criterion.

```python
import numpy as np

class MixLinearRegression:
    def __init__(self, prior=None, coef_=None, intercept_=None, sigma=None):
        """an initialization of the model, which could be extended
        with initial settings."""
        self.prior = prior
        self.sigma = sigma
        self.coef_ = coef_
        self.intercept_ = intercept_

    def fit(self, X, Y, K=1, run_min=50, run_max=100000, gap_ratio=0.0001, 
        is_stochastic=False):
        """implement the training"""
        # check the shape of the input
        if len(X.shape) == 1:
            X = X.reshape(-1,1)
        if len(Y.shape) == 2 and Y.shape[1] == 1:
            Y = Y.reshape(-1)
        self.X = X
        self.Y = Y
        X1 = np.ones((X.shape[0], X.shape[1]+1))
        X1[:,1:] = X
        N, M = X1.shape
        R = np.random.dirichlet(np.ones(K), N)
        Prior = np.ones(K)
        Coef  = np.ones((M, K))
        Sigma = np.ones(K)
        for i in range(run_max):
            #M-step
            Prior = R.mean(axis=0)
            if is_stochastic:
                for n in range(R.shape[0]):
                    R[n,:] = np.random.multinomial(1, R[n,:], size=1)
            for j in range(K):
                Coef[:,j] = np.linalg.inv(np.dot((X1*R[:,j:j+1]).T, X1)).dot((X1*R[:,j:j+1]).T).dot(Y)
                Sigma[j]  = np.sqrt(np.sum((Y - X1.dot(Coef[:,j]))**2 * R[:,j]) / np.sum(R[:,j]))
            #E-step
            for j in range(K):
                R[:,j] = Prior[j] / (Sigma[j] * np.sqrt(2*np.pi)) * np.exp(-0.5*(Y - X1.dot(Coef[:,j]))**2 / Sigma[j]**2)
            R[:,:] = R / np.sum(R, axis=1).reshape(-1,1)
            # check convergence
            loglik = 0
            for j in range(K):
                loglik += Prior[j] / (Sigma[j] * np.sqrt(2*np.pi)) * np.exp(-0.5*(Y - X1.dot(Coef[:,j]))**2 / Sigma[j]**2)
            loglik = np.sum(np.log(loglik))

            if i >= run_min and (loglik - loglik_old) < gap_ratio*abs(loglik):
                break
            else:
                loglik_old = loglik

        self.coef_ = Coef[1:,:]
        self.intercept_ = Coef[0,:]
        self.loglik = loglik
        self.sigma  = Sigma
        self.prior  = Prior
        self.probs  = R
        self.bic    = -2*loglik + (K*(M+2)-1)*np.log(N)
        
    def predict(self, Xtest):
        """implement the prediction"""
        if len(Xtest.shape) == 1:
            Xtest = Xtest.reshape(-1,1)
        RV = np.zeros(Xtest.shape[0])
        for j in range(self.coef_.shape[1]):
            RV += self.prior[j] * (Xtest.dot(self.coef_[:,j]) + 
                self.intercept_[j])
        return RV.reshape(-1)

    def predict_label(self, Xtest, Ytest):
        K = self.prior.shape[0]
        P = np.zeros(K, N)
        for j in range(K):
            P[:,j] = self.prior[j] * 1.0 / (self.sigma[j] * np.sqrt(2*np.pi))
            P[:,j] *= np.exp(-0.5*(Xtest.dot(self.coef_[:,j]) + 
                self.intercept_[j] - Ytest)**2 / self.sigma[j]**2)
        P[:,:] = P / np.sum(R, axis=1).reshape(-1,1)
        return P
```

```python
import scipy.stats as st
import pylab as pl
%matplotlib inline

X1 = np.random.rand(200, 5)
W1 = np.random.rand(5) * 20
Y1 = X1.dot(W1) + np.random.rand(200) * 5

X2 = np.random.rand(200, 5)
W2 = np.random.rand(5) * 20
Y2 = X2.dot(W2) + np.random.rand(200) * 5

X = np.append(X1, X2, axis=0)
Y = np.append(Y1, Y2)
print X.shape, Y.shape

# un-inversible
# X = np.array([[0, 0], [1, 1], [2, 2]])
# Y = np.array([0, 1, 2])

mixreg = MixLinearRegression()
mixreg.fit(X,Y,K=2)
Y_pre = mixreg.predict(X)
print mixreg.coef_, mixreg.intercept_, mixreg.sigma
# print W, np.mean(Y-X.dot(W)), np.std(Y-X.dot(W))

fig = pl.figure()
pl.plot(Y, Y_pre, "o", alpha=0.5)
pl.xlabel("observed $y$")
pl.ylabel("predicted $y$")
pl.title("Pearson's $R=%.3f$" %(st.pearsonr(Y, Y_pre)[0]))

# fig.set_size_inches(5,4)
# fig.savefig(data_dir = "test1.pdf", dpi=300, bbox_inches='tight')
pl.show()
```

```python
X = np.random.rand(400, 2)
X[:,1] = X[:,0] ** 2
W1 = np.array([-8, 8])
Y = X.dot(W1) + np.random.rand(400) * 0.7
W2 = np.array([6, -6])
Y[200:] = X[200:,:].dot(W2) + np.random.rand(200) * 0.5 - 2


mixreg = MixLinearRegression()
mixreg.fit(X,Y,K=2,is_stochastic=False)
Y_pre = mixreg.predict(X)
print mixreg.coef_, mixreg.intercept_, mixreg.sigma, mixreg.loglik

fig = pl.figure()
idx1 = mixreg.probs[:,0] > 0.5
idx2 = mixreg.probs[:,0] <= 0.5
pl.plot(X[idx1,0], Y[idx1], "bo", alpha=0.5)
pl.plot(X[idx2,0], Y[idx2], "ro", alpha=0.5)
pl.xlabel("observed $y$")
pl.ylabel("predicted $y$")
pl.title("Pearson's $R=%.3f$" %(st.pearsonr(Y, Y_pre)[0]))

# fig.set_size_inches(5,4)
# fig.savefig(data_dir = "test1.pdf", dpi=300, bbox_inches='tight')
pl.show()
```
